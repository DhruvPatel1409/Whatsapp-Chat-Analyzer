```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with _ A.I.T • I.T _(FRD).txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s[AP]M\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
from dateutil import parser

df = pd.DataFrame({'user_message': messages, 'message_date': dates})
df['message_date'] = df['message_date'].str.replace(',', ' ')
df['message_date'] = df['message_date'].str.replace('-', '')
df['message_date'] = pd.to_datetime(df['message_date'],format='mixed')
df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split('([^:]+):', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])  
    else:
        users.append('group_notification')
        messages.append(entry[0]) 

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
df['day_name'] = df['date'].dt.day_name()
```

```python
df['day_name'].value_counts()
```

```python
period = []
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
